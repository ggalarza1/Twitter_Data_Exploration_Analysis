```python
tweets_data = combined_tweets.drop(labels=["place", "geo", "near", "source", "user_rt_id", "user_rt", "retweet_id", "retweet_date","translate", "trans_src", "trans_dest", "timezone", "place", "quote_url", "search", "reply_to"], axis=1)
```

```python
cleaned_tweets =tweets_data.loc[:,['date', 'language', 'nlikes', 'tweet']]
```

```python
# Cleaning the tweets

def cleanUpTweet(txt):
    # Remove mentions
    txt = re.sub(r'@[A-Za-z0-9_]+', '', txt)
    # Remove hashtags
    txt = re.sub(r'#', '', txt)
    # Remove retweets:
    txt = re.sub(r'RT : ', '', txt)
    # Remove urls
    txt = re.sub(r'https?:\/\/[A-Za-z0-9\.\/]+', '', txt)
    return txt
```

```python
cleaned_tweets['tweet'] = cleaned_tweets['tweet'].apply(cleanUpTweet)
```

```python
def getTextSubjectivity(txt):
    return TextBlob(txt).sentiment.subjectivity

def getTextPolarity(txt):
    return TextBlob(txt).sentiment.polarity
```

```python
cleaned_tweets['Subjectivity'] = cleaned_tweets['tweet'].apply(getTextSubjectivity)
cleaned_tweets['Polarity'] = cleaned_tweets['tweet'].apply(getTextPolarity)
```

```python
cleaned_tweets.head(50)
```

```python
cleaned_tweets = cleaned_tweets.drop(cleaned_tweets[cleaned_tweets['tweet'] == ''].index)
```

```python
cleaned_tweets.head(50)
```

```python
# negative, nautral, positive analysis
def getTextAnalysis(a):
    if a < 0:
        return "Negative"
    elif a == 0:
        return "Neutral"
    else:
        return "Positive"
```

```python
cleaned_tweets['Score'] = cleaned_tweets['Polarity'].apply(getTextAnalysis)
```

```python
cleaned_tweets.head(50)
```

```python
positive = cleaned_tweets[cleaned_tweets['Score'] == 'Positive']

print(str(positive.shape[0]/(cleaned_tweets.shape[0])*100) + " % of positive tweets")
```

```python
labels = cleaned_tweets.groupby('Score').count().index.values

values = cleaned_tweets.groupby('Score').size().values

plt.bar(labels, values)
```

```python
for index, row in cleaned_tweets.iterrows():
    if row['Score'] == 'Positive':
        plt.scatter(row['Polarity'], row['Subjectivity'], color="green")
    elif row['Score'] == 'Negative':
        plt.scatter(row['Polarity'], row['Subjectivity'], color="red")
    elif row['Score'] == 'Neutral':
        plt.scatter(row['Polarity'], row['Subjectivity'], color="blue")

plt.title('Twitter Sentiment Analysis')
plt.xlabel('Polarity')
plt.ylabel('Subjectivity')
# add legend
plt.show()
```

```python
objective = cleaned_tweets[cleaned_tweets['Subjectivity'] == 0]

print(str(objective.shape[0]/(cleaned_tweets.shape[0])*100) + " % of objective tweets")
```

```python
# Creating a word cloud
words = ' '.join([tweet for tweet in cleaned_tweets['tweet']])
wordCloud = WordCloud(width=600, height=400).generate(words)

plt.imshow(wordCloud)
plt.show()
```
